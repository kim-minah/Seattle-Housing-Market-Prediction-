```python
import numpy as np
import pandas as pd
import requests
import os, sys
from bs4 import BeautifulSoup
```

```python
url = 'https://www.niche.com/places-to-live/search/zip-codes-with-the-best-public-schools/c/king-county-wa/'
useragent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
headers = {'user-agent': useragent}
r = requests.get(url, headers = headers)
r 
r.text # gives us html but python doesn't know that yet, need to parse it using BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')

#soup


# What prof does to find data:
# See if html is usable or not (is the data there somewhere to be extracted? do you need to use more invasive method like selenium)
# Right click -> View Page Source
# Search for title of conference, lives in a tag named <title> (tags have opening and closing tags)
# Use beautiful soup to extract tag and then extract actual title (navigable string)
# Tags can have attributes. In <meta name = "~" content = "~"> name and content are attributes
```

```python
soup.find_all("h2")
```

```python
zipcodes = pd.DataFrame([x.text for x in soup.find_all("h2") if len(x.text)<6])
zipcodes
zipcodes.to_csv('ranked_zipcodes.csv',sep=",",index=False)

#dat11.to_csv("dat11_cleaned.csv", sep=",",index=False)
```
